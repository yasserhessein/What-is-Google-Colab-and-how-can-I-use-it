# What is Google Colab and how can I use it ?


How to Make Use of Google Colab is the platform for you if you want to build a machine learning or  deep leraning model but don't have a computer that can handle the workload. Creating a local environment with Anaconda and downloading packages and fixing installation problems is a hassle even if you have a GPU or a decent device.


Google's Colaboratory is a free Jupyter notebook environment where you can use free GPUs and TPUs to solve all of these problems.



[



## The First Steps

To use Colab, first log in to your Google account, then go to

[LINK](https://colab.research.google.com.)



## The Second Step 


![](https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/uc1.png)


To make a new Colab notebook, click the NEW NOTEBOOK button. By clicking the upload button, you can also upload your local notebook to Colab:


![](https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/uc17.png)

It will create a Jupyter notebook with the name Untitled0.ipynb and save it to your Google Drive in a folder called Colab Notebooks when you create a new notebook. Since it's basically a Jupyter notebook, all Jupyter notebook commands will work.



![](https://media.geeksforgeeks.org/wp-content/uploads/20190430120025/Screenshot-4012.png)

## The Third Step 

Choosing the GPU or TPU Option

Select Change runtime type. Now select anything(GPU, CPU, None) you want in the Hardware accelerator dropdown menu.



## GPUs and TPUs


It provides you with a good GPU for free, which you can use for up to 12 hours. For most data scientists, this is enough to satisfy their computation requirements. I would strongly advise you to begin using Google Colab, particularly if you are a beginner.




For our notebooks, Google Colab provides three forms of runtime:


* CPUs,
* GPUs, and
* TPUs

![](https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/uc2-e1583900164853.png)


## The Fourth Stup 

Uploading Files and Datasets

Importing your dataset into Colab is the first step on your data analysis journey.


The most straightforward method is to upload your data directly to Colab:

![](https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/uc7.png)

Since the upload speed in this method is very slow, you can use it if your dataset or file is quite tiny. Another choice is to upload your dataset to Google Drive and then instal the drive on Colab. You can do this with only one mouse click:


![](https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/uc8.png)

## The Fifth Step


Downlonad Your Notebook

All of your Colab notebooks are saved to your Google Drive. The best part about Colab is that your notebook is automatically saved after a certain amount of time has passed, so you won't lose your work.

You can export and save your notebook in both *.py and *.ipynb formats if you want to:

![](https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/uc13.png)
